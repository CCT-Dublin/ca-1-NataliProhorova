```python
# identifying columns with missing data and deciding how to handle  missing values
print(df.isnull().sum())
```

```python
#assuming 222 missed values in a column "Gender",replase them with 'Not applicable' or 'NA'

df['Gender'] = df['Gender'].fillna('NA')
```

```python
# identifying  categories/values in a  column
df['Gender'].unique()
```

```python
#Replace with "NA" (No answer) non gender related  categories  using the replace() method:
df['Gender'] = df['Gender'].replace(['Prefer not to answer', 'Other or prefer not to answer'], 'NA')
```

```python
#identifying unique duplicates in a columns
df.drop_duplicates('Gender', keep='first')
```

```python
#  determine unique categories in Higest  Education column
array = df["highest education level"].unique() 
print(array)
```

```python
#replace missing values with 'NA'  (no answer)
df['highest education'] = df['highest education'].fillna('NA')
```

```python
#rename values
df['highest education'] = df['highest education'].replace(['Professional degree (MD, JD, etc.)','Master\'s degree' ], 'MD')
```

```python
#rename values
df['highest education'] = df['highest education'].replace(['Some college','College degree'], 'CD')
```

```python
# Determine the distribution of categorical variables 
# Frequency table for 'highest education'
Education_distribution = df['highest education'].value_counts()

print(Education_distribution)
```
